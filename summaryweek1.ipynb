MaxElement

```python
def MmxElement(A):
    maxval = A[0]
    for i in range(1, len(A)):
        if A[i] > maxval:
            maxval = A[i]
    return maxval
```

```python
list1 = [3, 5, 2, 8, 1]
print(MmxElement(list1))
```

Euclid's algorithm

```python
def Euclid (m, n):
    while n != 0:
        r = m % n
        m = n
        n = r
    return m
```

```python
Euclid(0, 18)
```

Sieve of Eratosthenes

```python
def sieve_of_eratosthenes(n):
    A = []
    if n >= 2:
        for p in range(2, n + 1):
            A.append(p)
        for p in range(2, int(n**0.5) + 1):
            if p in A:
                for j in range(p * p, n + 1, p):
                    if j in A:
                        A.remove(j)
    return A
```

```python
sieve_of_eratosthenes(25)
```

Sum All Element

```python
def Sum_All_Elements(A):
    sum = 0
    for i in range(len(A)):
        sum += A[i]
    return sum
```

```python
Sum_All_Elements([1, 2, 3, 4, 5])
```

Unique Element

```python
def Unique_Elements(A):
    for i in range(len(A)):
        for j in range(i + 1, len(A)):
            if A[i] == A[j]:
                return False
    return True
```

```python
Unique_Elements([1,2,3,3])
```

Factorial

```python
def Factorial(n):
    if n == 0:
        return 1
    else:
        return Factorial(n - 1)* n
```

```python
Factorial(5)
```

Fibonacci

```python
def Fibonacci(n):
    if n <= 1:
        return n
    else:
        return Fibonacci(n - 1) + Fibonacci(n - 2)
```

```python
Fibonacci(6)
```
