### Bubble Sort

```python
#Algorithm BubbleSort (A [0..n-1])  
# // Sorts a given array by bubble sort  
# Input: An array A[0..n-1] of orderable elements  
# Output: Array A[0..n-1] sorted in nondecreasing order  

# 1 for i = 0 to n-2 do  
# 2    for j = 0 to n-2-i do  
# 3            if A[j+1] < A[j]      
# 4                     swap A[j] and A[j+1]  
```

<span style="color:red">Write your code here! </span>

```python
def BubbleSort(A):
    n = len(A)
    for i in range(n - 1):
        for j in range(n - 1 - i):
            if A[j + 1] < A[j]:
                A[j], A[j + 1] = A[j + 1], A[j]
    return A    
 
```

```python
def create_a_sequence(a,b,n):
    from random import randint
    A = []
    for i in range(n):
        A.append(randint(a,b))
    print(A)
    return A
```

### Selection Sort

```python
#Algorithm SelectionSort ( A[0..n-1] )
#// Sorts a given array by selection sort
#Input: An array A[0..n-1] of orderable elements
#Output: Array A[0..n-1] sorted in nondecreasing order

#1. for i = 0 to n-2 do
#2.      min = i
#3.      for j = i+1 to n-1 do
#4.          if A[j] < A[min]    
#5.                 min = j
#6.      swap A[i] and A[min]
```

<span style="color:red">Write your code here! </span>

```python
def SelectionSort(A):
    n = len(A)
    for i in range(n - 1):
        min = i
        for j in range(i + 1, n):
            if A[j] < A[min]:
                min = j
        A[i], A[min] = A[min], A[i]
    return A
```

Create a random-interger sequence and Test your code

```python
A = create_a_sequence(1,100,10)
```

```python
SelectionSort(A)
```

### Brute Force String Matching

```python
def BruteForceStringMatching(P,T):
###Return the lowest index of T at which substring P begins or -1 for not found.
    n = len(T)
    m = len(P)
    for i in range(n - m + 1):
        j = 0
        while j < m and P[j] == T[i + j]:
            j += 1
        if j == m:
            return i
    return -1
```

### Closest-Pair Problem

AlternativeLogo-eng.png

```python
import math
from math import sqrt
from random import seed,randint

#Euclidean distance
def dist(a,b):
    x=pow((a[0]-b[0]),2)
    y=pow((a[1]-b[1]),2)
    return sqrt(x+y)


def BruteForceClosetPair(P):
    n=len(P)
    d=math.inf
    for i in range(n-1):
       for j in range(i+1,n):
           d=min(d,dist(P[i],P[j]))
    return d
```

Test your code

```python
#generate tuple of points
seed(14)
max = 1000
num_points = 500
points = [(randint(0, max), randint(0, max)) for _ in range(num_points)]
print(points)
```

```python
BruteForceClosetPair(points)
```
